## Manipulasi Vektor/Matriks; Aljabar Linear (poin= 30)

1. Buat definisi Matriks M dan Vektor a, b, c.  
Salah satu package Python yang umum digunakan untuk permasalahan terkait aljabar linear seperti ini adalah Numpy.

```python
M = Matrix([[1, 2, 3], [4, 5, 6], [7, 8, 9], [0, 2, 2]])
M
```

```python
a = Matrix([[1], [1], [0]])
b = Matrix([[-1], [2], [5]])
c = Matrix([[0], [2], [3], [2]])
a, b, c
```

2. Hitung dot product dari vektor a dan b.  
Simpan nilainya ke dalam variabel aDotb dan tampilkan nilainya ke layar.

```python
aDotb = a.dot(b)
aDotb
```

3. Hitung _element-wise product_ dari vektor a dan b.

```python
a.multiply_elementwise(b)
```

4. Hitung (a<sup>T</sup>b)Ma.  
Tampilkan hasilnya ke layar.

```python
aTb = a.T*b
aTb
```

```python
Ma = M*a
Ma
```

```python
aTb*Ma
```

> Syntax diatas akan menghasilkan error _Matrix size mismatch: (1, 1) * (4, 1)_.  
Ini karena hasil perhitungan _aTb_ menghasilkan matriks 1x1, sehingga tidak dapat dikalikan dengan matrik 4x1. Kecuali hasil dari perkalian _aTb_ yang berukuran 1x1 dijadikan nilai skalar (_dot product_).

```python
aTbS = np.matrix(aTb)
aTbS = np.asscalar(aTbS)
aTbS
```

```python
aTbS*Ma
```

5. Tanpa menggunakan _loop_, hitung perkalian _element-wise_ dari tiap baris M dengan a.  
Tampilkan hasilnya ke layar.  
Petunjuk: bisa menggunakan fungsi __repmat()__.

```python
no_loop1 = np.multiply(npm.repmat(a.T, 4, 1), np.asarray(M))
Matrix(no_loop1)
```

6. Tanpa menggunakan _loop_, urutkan semua nilai dari M yang baru (dari hasil 5) terurut dari nilai terkecil ke terbesar lalu buat dalam grafik/plot.

## Basic image manipulation (poin=70)

1. Read in the images, image1.jpg and image2.jpg, as color images.

```python
img1 = plt.imread('image1.jpg')
img2 = plt.imread('image2.jpg')
```

```python
f, (ax1, ax2) = plt.subplots(1, 2)

ax1.imshow(img1)
ax1.set_title('image1.jpg')
ax1.set_axis_off()

ax2.imshow(img2)
ax2.set_title('image2.jpg')
ax2.set_axis_off()
```

2. Convert the images to double precision and rescale them to stretch from minimum value 0 to maximum value 1.

```python
img1_dblp=skimage.img_as_float64(img1)
img2_dblp=skimage.img_as_float64(img2)
```

3. Add the images together and re-normalize them to have minimum value 0 and maximum value 1. Display this new image.

```python
img_add = img1_dblp + img2_dblp
img_add = img_add/(img_add.max())

f, ax = plt.subplots(1, 1)
ax.imshow(img_add)
ax.set_title('Adding Images')
ax.set_axis_off()
```

4. Create a new image such that the left half of the image is the left half of image1 and the right half of the image is the right half of image2. Display this new image.

```python
img1_left_half = img1[0:500, 0:250, :]
img1_right_half = img1[0:500, 250:500, :]
img2_left_half = img2[0:500, 0:250, :]
img2_right_half = img2[0:500, 250:500, :]

img_combine1 = np.concatenate((img1_left_half, img2_right_half), axis=1)
img_combine2 = np.concatenate((img2_left_half, img1_right_half), axis=1)

f, (ax1, ax2) = plt.subplots(1, 2)

ax1.imshow(img_combine1)
ax1.set_title('Combination No. 1')
ax1.set_axis_off()

ax2.imshow(img_combine2)
ax2.set_title('Combination No. 2')
ax2.set_axis_off()
```

> This can be done because the image size are equal with each other. If the image size are different, then it needs to be rescaled to the same size first.

5. Using a for loop, create a new image such that every odd numbered row is the corresponding row from image1 and the every even row is the corresponding row from image2. (Hint: Remember that indices start at 0 and not 1 in Python). Display this new image.

```python
img_odd_even = np.ones((500, 500, 3))
for i in range(img_odd_even.shape[0]):
    if i%2 == 0:
        img_odd_even[i, :, :] = img1_dblp[i, :, :]
    else:
        img_odd_even[i, :, :] = img2_dblp[i, :, :]

f, ax = plt.subplots(1, 1)
ax.imshow(img_odd_even)
ax.set_title('Odd-Even Image For-Loop')
ax.set_axis_off()
```

6. Accomplish the same task as part 5 without using a for-loop (the functions reshape and repmat may be helpful here).

```python
img_odd_even_no_loop = np.ones((500, 500, 3))

img_odd_even_no_loop[::2] = img1_dblp[::2]
img_odd_even_no_loop[1::2] = img2_dblp[1::2]

f, ax = plt.subplots(1, 1)
ax.imshow(img_odd_even_no_loop)
ax.set_title('Odd-Even Image No Loop')
ax.set_axis_off()
```

7. Convert the result from part 6 to a grayscale image. Display the grayscale image with a title.

```python
img_grey = skimage.color.rgb2gray(img_odd_even_no_loop)
```

```python
f, ax = plt.subplots(1, 1)
ax.imshow(img_grey, cmap='gray')
ax.set_title('Grayscale Image')
ax.set_axis_off()
```
